```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = 'GoodDataset-1.csv'

df = pd.read_csv(f'Dataset/{dataset}')

with pd.option_context('display.max_rows', 6): display(df)

print(df.shape)

class_count = df['Class'].value_counts()
print(class_count)

plt.bar(['Low Sweetness', 'High Sweetness'], [class_count[0], class_count[1]])
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
n = 500
threshold = 18
augmented_data = []
column = ['Weight (g)', 'Roundness', 'Lower Petal', 'Max Frequency (Hz)', 'Max Magnitude', 'Sweetness (%Brix)']

for _ in range(n):
    selected_row = df.iloc[np.random.randint(0, len(df))]
    data = selected_row[column]
    
    noise = np.random.normal(0, 2.0, len(data))
    noise[1], noise[2] = 0, 0
        
    augmented = data + noise
    augmented['Roundness'] = round(np.random.uniform(0.912, 0.981), 3)
    augmented['Lower Petal'] = np.random.randint(5, 8)
        
    augmented_data.append(augmented)
      
augmented_data = pd.DataFrame(augmented_data)
display(augmented_data)
```

```python
new_df = pd.concat([df, augmented_data])

with pd.option_context('display.max_rows', None): display(new_df)
    
print(new_df.shape)
```

```python
new_df.to_csv(f'AugmentedDataset-2.csv', index=False)
```
